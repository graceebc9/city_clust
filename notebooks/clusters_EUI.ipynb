```python
df = wrkv1.copy() 

df=df[df['percent_residential']==1]
print('Len on inut data is ', len(df))
df = df.merge(input_cl, on='TCITY15NM')
```

```python
# /group by cluster and get the mean and standrad deviation of the eui
df = df.groupby('cluster')['eui'].agg(['mean', 'std']).reset_index()
```

```python
# plot the box plot of eui for each cluster
import seaborn as sns
import matplotlib.pyplot as plt
def boxplot_cluster(input_cl, label_path, df ):
    
    labels  = pd.read_csv(label_path)['Cluster']
    input_cl['cluster'] = labels


    df=df[df['percent_residential']==1]
    print('Len on inut data is ', len(df))
    df = df.merge(input_cl, on='TCITY15NM')


    sns.boxplot(x='cluster', y='eui', data=df)
```
